```python
from IPython.display import Latex
from PyFoam.RunDictionary.SolutionDirectory import SolutionDirectory
from PyFoam.RunDictionary.ParsedParameterFile import ParsedParameterFile
from PyFoam.Basics.DataStructures import Vector
import numpy as np
import pandas as pd
import os,shutil,glob
import pylab, re
import matplotlib.pyplot as plt
from tikzplotlib import save as tikz_save
import tikzplotlib as tikzpl
import matplotlib as mpl
import matplotlib.font_manager
from matplotlib.backends.backend_pgf import FigureCanvasPgf
mpl.backend_bases.register_backend('pdf', FigureCanvasPgf)
plt.rcParams.update({
    "pgf.rcfonts": False,
    "font.family": "serif",  # use serif/main font for text elements
    "font.serif": ['Tex Gyre Pagella'],
    "text.usetex": True,     # use inline math for ticks
    'pgf.texsystem': 'lualatex',
    "pgf.preamble": "\n".join([
        r"\PassOptionsToPackage{osf,sc}{mathpazo}",
        r"\RequirePackage{mathpazo}",
        r"\usepackage{pgfplots}",
        r"\usepgfplotslibrary{external}",
        r"\tikzexternalize",
    ])
})
```

```python
for i in range (0,8):
    subA=pd.read_csv('/media/andres/LFD/andres/newDomain/V083/10Hz/fluid/data/AvgField{}.csv'.format(i),skiprows=1)
    #print(subA)
    appA=subA.to_numpy()
    #print(appA)
    if i==0:
        A=subA.to_numpy()
    else:
        A=np.append(A,appA,0)
data=A
f=10 #Hz
fplus=f/13.27 #Hz
```

```python
#Obtengo la magnitud de la velocidad
magU=np.sqrt(np.power(data[:,2],2)+np.power(data[:,3],2))
#Adimensionalizo con la velocidad de freestream
normU=magU/0.83
#Extraigo coordenadas, les resto la coordenada del centro de rotacin y adimensionalizo con L.
x=(data[:,18]-0.031)/0.035
y=(data[:,19]-0.03)/0.035
#Creo un vector con los distintos conjuntos de nivel.
levels=np.linspace(normU.min(),normU.max(),200)
#Setteamos LaTeX en los plots. Usamos Palatino, la fuente de la tesis.
plt.rcParams['text.usetex'] = True
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
#Creamos la figura
fig, ax = plt.subplots(figsize=(8,3))
Campo = plt.tricontourf(x,y,normU,levels=levels) #El tricontourf
for c in Campo.collections:
    c.set_edgecolor("face") #Sacamos las lneas blancas que mete de los conjuntos de nivel y nos arruinan el pdf.
#Setteamos los lmites de los ejes - Iguales a los del paper.
ax.set(xlim=(0, 8), ylim=(-1.5, 1.5)) 
ax.set_xlabel(r'$z_1/L$',size=12)
ax.set_ylabel(r'$z_2/L$',size=12)
ax.set_title('$V_R=0.28$',loc='left',size=12)
ax.set_title('$f^+={:0.2f}$'.format(fplus),loc='right',size=12)
#Un poco de formato.
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
#Definimos la colorbar y los ticks a meter.
CB=plt.colorbar(Campo, ticks=np.linspace(0,normU.max(),8))
CB.ax.tick_params(labelsize=12)
CB.ax.set_ylabel(r'$\displaystyle\frac{\langle\|\underline{\mathbf{v}}\|\rangle}{V_{\infty}}$',rotation=0, size=12, labelpad=20)
#Dibujamos la grilla
plt.grid() 
ax.grid(axis='both',which='both',color='black', linestyle=':', linewidth=0.5,zorder=1000)
#Guardamos la figura
plt.tight_layout()
plt.savefig('PromediomagnitudV-083ms-10Hz.pdf') 
plt.savefig('PromediomagnitudV-083ms-10Hz.png')  
```
